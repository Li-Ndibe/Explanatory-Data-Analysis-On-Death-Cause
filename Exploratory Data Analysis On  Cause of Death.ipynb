```python
death_cause = ['Meningitis', 'Dementias',
       'Parkinson_Disease', 'Nutritional_Deficiencies', 'Malaria', 'Drowning',
       'Interpersonal_Violence', 'Maternal_Disorders', 'HIV_AIDS',
       'Drug_Use_Disorders', 'Tuberculosis', 'Cardiovascular_Diseases',
       'Lower_RespiratoryInfections', 'Neonatal_Disorders',
       'Alcohol_Disorders', 'Self_harm', 'Nature_Forces', 'Diarrheal_Diseases',
       'Environmental_Exposure', 'Neoplasms', 'Conflict', 'Diabetes_Mellitus',
       'Kidney_Disease', 'Poisonings', 'Protein_Malnutrition', 'Road_Injuries',
       'Respiratory_Diseases', 'Liver_Diseases', 'Digestive_Diseases',
       'Hot_Substances', 'Acute_Hepatitis']
```

```python
# Creating a new column for 'death_sum' for individual Country and Year

cd['death_sum'] = cd[death_cause].sum(axis=1)
```

```python
# Top 10 Deaths sum

top10_death_sum = cd.sort_values(by='death_sum',ascending=False)[:10][['death_sum','Country']]

top10_death_sum
```
